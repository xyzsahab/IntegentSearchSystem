```python
def search_paper(query,k=5):
  query_embedding=model.encode([query])

  faiss.normalize_L2(query_embedding)

  D,I=index.search(query_embedding,k)
  count=1;
  for score, idx in zip(D[0],I[0]):
    print("Result:",count)
    count+=1
    print("\tSimilarity score", score)
    print("\tTitle: ", df.iloc[idx]["title"])
    print("\tabstract: ",df.iloc[idx]["abstract"])
```

```python
# Step 4 setup: load a zero-shot classification model to judge query difficulty
# (uses the same transformers library as the summarizer above)

from transformers import pipeline

difficulty_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# Step 4: Difficulty Level
# We ask the zero-shot model how well the query matches each difficulty label

def check_difficulty(query):
  labels = ["beginner", "intermediate", "advanced"]

  result = difficulty_classifier(query, candidate_labels=labels)

  level = result["labels"][0].capitalize()

  reasons = []
  for label, score in zip(result["labels"], result["scores"]):
    reasons.append(label.capitalize() + " score: " + str(round(score, 2)))

  return level, reasons
```

```python
# Step 5: Related Searches
# We reuse the FAISS search results and run KeyBERT on each matched abstract
# to pull out real keywords from the corpus, instead of a hand made list

def get_related_searches(result_indexes):
  related = []

  for idx in result_indexes:
    abstract = df.iloc[idx]["abstract"]
    keywords = kw_model.extract_keywords(abstract, top_n=3)

    for word, score in keywords:
      if word not in related:
        related.append(word)

  return related[:6]
```

```python
# Full flow: Search -> Summarization -> Keyword Extraction -> Difficulty -> Related Searches

def full_search_query(query, k=5):
  # Step 1: Search
  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)
  D, I = index.search(query_embedding, k)
  top_idx = I[0][0]
  top_abstract = df.iloc[top_idx]["abstract"]

  # Step 2: Summarization
  summary = summarizer(top_abstract, max_length=120, min_length=40, do_sample=False)
  summary_text = summary[0]["summary_text"]

  print("Summary")
  print(summary_text)
  print()

  # Step 3: Keyword Extraction
  keywords = kw_model.extract_keywords(top_abstract)
  print("Keywords")
  for word, score in keywords:
    print("-", word)
  print()

  # Step 4: Difficulty Level
  level, reasons = check_difficulty(query)
  print("Difficulty")
  print(level)
  print()
  print("Reason")
  for reason in reasons:
    print("-", reason)
  print()

  # Step 5: Related Searches (built from the actual top search results)
  related = get_related_searches(I[0])
  print("Related Searches")
  for item in related:
    print("-", item)
```

```python
full_search_query("medical research python fastapi")
```
